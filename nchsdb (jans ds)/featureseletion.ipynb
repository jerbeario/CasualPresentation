```python
import pandas as pd
```

```python
df = pd.read_csv("data/interim.csv")
data_dict = pd.read_csv("data/answers-data-dictionary-0.2.0-domains.csv", sep=";")
```

```python
df.isna().sum()[df.isna().sum() > 0].sort_values(ascending=True)
```

```python
df = df.drop(["id", "visit", "msfsc2"], axis=1)
df.dropna(subset=["education", "age"], inplace=True)
```

```python
df.isna().sum()[df.isna().sum() > 0].sort_values(ascending=True)
```

```python
cols_for_total_score = [col for col in df.columns if "total" in col]
cols_for_total_score
```

```python
cols_lifestyle = ["subs_alc_1", "subs_tob_1", "subs_can_1"]
```

```python
cols_sleep = ["socialjetlag", "msfsc1"]
```

```python
cols_medical_history = data_dict[(data_dict['folder'].str.contains("Medical History")) & (data_dict['folder'].str.contains("Medications") == False)].id.to_list()
cols_medical_history
```

```python
demographic_cols = data_dict[(data_dict['folder'].str.contains("Demographics")) & (data_dict['id'].str.contains("nsrr") == False)].id.to_list() + ["handedness"]
demographic_cols
```

```python
# df[cols_for_total_score]
df[cols_lifestyle]
df[cols_sleep]
df[cols_medical_history]
```

```python
selected_cols = cols_lifestyle + cols_sleep + cols_medical_history + demographic_cols
selected_df = df[selected_cols]
```

```python
selected_df.info()
```

```python
for column in selected_df.columns:
    if selected_df[column].dtype == "object":
        print(column, selected_df[column].unique())
```

```python
subs_alc_1_map = {
    'Never': 0,
    'Once a month or less': 1,
    'Once a week or less': 2,
    'A few times a week': 3,
    'Daily': 4,
    'Multiple times daily': 5
}

subs_tob_1_map = {
    'Never Smoker': 0,
    'Former Smoker': 1,
    'Occasional Smoker': 2,
    'Daily Smoker': 3
}

subs_can_1_map = {
    'Never': 0,
    'Once a month or less': 1,
    'Once a week or less': 2,
    'A few times a week': 3,
    'Daily': 4,
    'Multiple times daily': 5
}

mh_anx_map = {
    'Not sure': -1,
    'No': 0,
    'Yes but undiagnosed': 1,
    'Yes diagnosed by clinician': 2
}

mh_bpd_map = {
    'Not sure': -1,
    'No': 0,
    'Yes but undiagnosed': 1,
    'Yes diagnosed by clinician': 2
}

mh_health_map = {
    'Poor': 0,
    'Fair': 1,
    'Good': 2,
    'Very Good': 3,
    'Excellent': 4
}

mh_mdd_map = {
    'Not sure': -1,
    'No': 0,
    'Yes but undiagnosed': 1,
    'Yes diagnosed by clinician': 2
}

mh_ptsd_map = {
    'Not sure': -1,
    'No': 0,
    'Yes but undiagnosed': 1,
    'Yes diagnosed by clinician': 2
}

mh_pzd_map = {
    'Not sure': -1,
    'No': 0,
    'Yes but undiagnosed': 1,
    'Yes diagnosed by clinician': 2
}

mh_szd_map = {
    'Not sure': -1,
    'No': 0,
    'Yes but undiagnosed': 1,
    'Yes diagnosed by clinician': 2
}

education_map = {
    "High school or less": 0,
    "College or more": 1
}

income_map = {
    "< $25,000": 0,
    "$25,000 - $50,000": 1,
    "$50,000 - $100,000": 2,
    "$100,000 - $200,000": 3,
    ">$200,000": 4,
    "Prefer not to say": -1
}

ethnicity_map = {
    "Non-Hispanic": 0,
    "Hispanic": 1
}

orientation_map = {
    "Heterosexual": 0,
    "Bisexual": 1,
    "Homosexual": 2,
    "Other": 3
}

race_map = {
    "White": 0,
    "Multiracial": 1,
    "Asian": 2,
    "Native": 3,
    "Black": 4,
    "Islander": 5
}

sex_map = {
    "Female": 0,
    "Male": 1
}

transgender_map = {
    'No': 0,
    'Yes': 1
}

handedness_map = {
    'Right': 0,
    'Left': 1,
    'Both': 2
}

selected_df["subs_alc_1"] = selected_df["subs_alc_1"].map(subs_alc_1_map)
selected_df["subs_tob_1"] = selected_df["subs_tob_1"].map(subs_tob_1_map)
selected_df["subs_can_1"] = selected_df["subs_can_1"].map(subs_can_1_map)
selected_df["mh_anx"] = selected_df["mh_anx"].map(mh_anx_map)
selected_df["mh_bpd"] = selected_df["mh_bpd"].map(mh_bpd_map)
selected_df["mh_health"] = selected_df["mh_health"].map(mh_health_map)
selected_df["mh_mdd"] = selected_df["mh_mdd"].map(mh_mdd_map)
selected_df["mh_ptsd"] = selected_df["mh_ptsd"].map(mh_ptsd_map)
selected_df["mh_pzd"] = selected_df["mh_pzd"].map(mh_pzd_map)
selected_df["mh_szd"] = selected_df["mh_szd"].map(mh_szd_map)
selected_df["education"] = selected_df["education"].map(education_map)
selected_df["income"] = selected_df["income"].map(income_map)
selected_df["ethnicity"] = selected_df["ethnicity"].map(ethnicity_map)
selected_df["orientation"] = selected_df["orientation"].map(orientation_map)
selected_df["race"] = selected_df["race"].map(race_map)
selected_df["sex"] = selected_df["sex"].map(sex_map)
selected_df["transgender"] = selected_df["transgender"].map(transgender_map)
selected_df["handedness"] = selected_df["handedness"].map(handedness_map)
```

```python
selected_df.drop(["mh_szd", "mh_pzd", "mh_ptsd", "mh_bpd", "handedness", "ethnicity", "subs_can_1"], axis=1, inplace=True)
```
